```python
import pandas as pd
import requests

def get_lat_long(api_key, address):
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    params = {
        "address": address,
        "key": api_key
    }

    try:
        response = requests.get(base_url, params=params)
        data = response.json()
        if data['status'] == 'OK':
            location = data['results'][0]['geometry']['location']
            lat = location['lat']
            lng = location['lng']
            return lat, lng
        else:
            print("Error:", data['status'])
            return None
    except Exception as e:
        print("An error occurred:", e)
        return None

api_key = 'AIzaSyD69p6gzMDgCygQzpk8LaBGyKM-_6znbDw'

cData['Latitude'] = None
cData['Longitude'] = None

for index, row in cData.iterrows():
    address = row['Incident place']
    if pd.notna(address):
        lat_lng = get_lat_long(api_key, address)
        if lat_lng:
            cData.at[index, 'Latitude'], cData.at[index, 'Longitude'] = lat_lng

print(cData.head())
```
